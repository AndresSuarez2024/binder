```python
import pandas as pd
```

```python
# Carregar el dataset des de la teva ubicació local
df = pd.read_csv('airquality.csv')
```

```python
# Mostrar les 10 primeres línies
print(df.head(10))
```

```python
# Mostrar la longitud del dataset (nombre de files)
print(f"La longitud del dataset és: {df.shape[0]}")
```

```python
# Mostrar les característiques de cada columna (tipus de dada, valors no nuls, etc.)
df.info()
```

```python
# Trobar la temperatura màxima
max_temp = df['Temp'].max()

# Imprimir la temperatura màxima
print(f"La temperatura màxima és: {max_temp}")
```

```python
# Trobar la fila on la temperatura és màxima
max_temp_row = df[df['Temp'] == df['Temp'].max()]

# Mostrar el mes i el dia corresponent a la temperatura màxima
max_temp_month_day = max_temp_row[['Month', 'Day']]
print(f"El mes i el dia amb la temperatura màxima són:\n{max_temp_month_day}")
```

```python
import matplotlib.pyplot as plt

# Crear el gràfic de punts (scatterplot)
plt.figure(figsize=(8,5))
plt.scatter(df['Wind'], df['Solar.R'], color='blue', alpha=0.6)

# Afegir etiquetes als eixos i títol
plt.xlabel("Velocitat del Vent")
plt.ylabel("Radiació Solar")
plt.title("Relació entre Velocitat del Vent i Radiació Solar")

# Mostrar el gràfic
plt.show()
```

```python
# interpreteu aquest gràfic
El gráfico muestra que no hay una relación clara entre la velocidad del viento y la radiación solar
```

```python
import matplotlib.pyplot as plt

# Calcular la temperatura mitjana per cada mes
temperature_per_month = df.groupby('Month')['Temp'].mean()

# Crear el gràfic de barres
plt.figure(figsize=(10, 6))
temperature_per_month.plot(kind='bar', color='skyblue', edgecolor='black')

# Afegir títols i etiquetes
plt.title("Temperatura Mitjana per Mes", fontsize=14)
plt.xlabel("Mes", fontsize=12)
plt.ylabel("Temperatura Mitjana (°F)", fontsize=12)
plt.xticks(rotation=0)  # Rotar les etiquetes del mes per facilitar la lectura

# Mostrar el gràfic
plt.show()
```

```python
# interpreteu aquest gràfic
El gráfico muestra que la temperatura media aumenta de mayo a julio-agosto, alcanzando su punto más alto
```

```python
import matplotlib.pyplot as plt

# Crear el gràfic de punts (scatterplot) per veure la relació entre Ozó i Temperatura
plt.figure(figsize=(8, 5))
plt.scatter(df['Ozone'], df['Temp'], color='green', alpha=0.6)

# Afegir títols i etiquetes
plt.title("Relació entre Ozó i Temperatura", fontsize=14)
plt.xlabel("Concentració d'Ozó", fontsize=12)
plt.ylabel("Temperatura (°F)", fontsize=12)

# Mostrar el gràfic
plt.show()
```

```python
# interpreteu aquest gràfic
El gráfico muestra que cuando el ozono aumenta, la temperatura también sube.
```
